```python
import os
from pathlib import Path

root = Path.cwd()
if not (root / "dataset").exists() and (root.parent / "dataset").exists():
    root = root.parent

train_dir = root / "dataset" / "part1" / "train"
masked_dir = train_dir / "masked"
unmasked_dir = train_dir / "unmasked"

if not masked_dir.exists() or not unmasked_dir.exists():
    raise FileNotFoundError(f"Dataset folders not found: {masked_dir} / {unmasked_dir}")

masked = set(os.listdir(masked_dir))
unmasked = set(os.listdir(unmasked_dir))
print(f"Masked: {len(masked)}, Unmasked: {len(unmasked)}")
print(f"Paired: {len(masked & unmasked)}")
print(f"Sample pairs: {list(masked & unmasked)[:5]}")
```
